```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Teen_Mental_Health_Dataset.csv')
```

```python
print(df.head())
print(df.info())
```

```python
print(df.describe().T)
```

```python
sns.set_style("whitegrid")
```

```python
# histogram - how many hours do teens spend on social media?
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["daily_social_media_hours"], bins=10, color="skyblue", edgecolor="black")
plt.title("Daily Social Media Use")
plt.xlabel("Number of day")
plt.ylabel("Number of teens")
plt.savefig("Chart1_social_media_hours.png")
plt.show()
```

```python
# bar chart - average stress level by social interaction level
```

```python
plt.figure(figsize=(8,5))
avg_stress = df.groupby("social_interaction_level")["stress_level"].mean()
avg_stress.plot(kind="bar", color="salmon")
plt.title("Average Stress Level by Social Interaction")
plt.xlabel("Social Interaction Level")
plt.ylabel("Average stress ()-10)")
plt.xticks(rotation=0)
plt.savefig("Chart2_stress_by_interaction.png")
plt.show()
```

```python
# scatter plot - sleep hours vs stress level
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["sleep_hours"], df["stress_level"], alpha=0.5, color="green")
plt.title("Sleep Hours vs Stress Level")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.savefig("Chart3_sleep_vs_stress.png")
plt.show()
```

```python
# boxplot - Anxiety level by gender
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="gender", y="anxiety_level")
plt.title("Anxiety Level by Gender")
plt.xlabel("Gender")
plt.ylabel("Anxiety Level (0-10)")
plt.savefig("chart4_anxiety_by_gender.png")
plt.show()
```

```python
# correlation heatmap - how are variables related?
```

```python
plt.figure(figsize=(8, 6))
numeric_columns = df.select_dtypes(include="number")
correlation = numeric_columns.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Variables")
plt.savefig("chart5_correlation_heatmap.png")
plt.show()
 
print("All charts saved as PNG files!")
 
```
